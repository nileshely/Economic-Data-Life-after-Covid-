```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/kaggle/input/economic-data-life-after-covid/economic data.csv")
```

```python
df.head(2)
```

```python
# 3D Scatter Plot of High, Low, and Close Prices
fig = px.scatter_3d(df, x='high', y='low', z='close', color='country')
fig.update_layout(title='3D Scatter Plot of High, Low, and Close Prices', scene=dict(xaxis_title='High', yaxis_title='Low', zaxis_title='Close'))
fig.show()
```

```python
# Histogram of Country Distribution
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='country', data=df, palette='viridis')
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')

plt.show()
```

```python
# Bar Plot of Services PMI by Country
plt.figure(figsize=(10, 6))
sns.barplot(x='country', y='services pmi', data=df, palette='muted')
plt.title('Services PMI by Country')
plt.xlabel('Country')
plt.ylabel('Services PMI')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot of Consumer Confidence
plt.figure(figsize=(8, 6))
sns.boxplot(y='consumer confidence', data=df, color='orange')
plt.title('Consumer Confidence Distribution')
plt.ylabel('Consumer Confidence')
plt.show()
```

```python
# Pairplot of Select Economic Indicators
sns.pairplot(df[['manufacturing pmi', 'services pmi', 'consumer confidence', 'unemployment rate']], height=3)
plt.suptitle('Pairplot of Economic Indicators')
plt.show()
```

```python
# Distribution Plot of Ticker Open Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['open'], kde=True, color='purple')
plt.title('Distribution of Ticker Open Prices')
plt.xlabel('Open Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Violin Plot of High Prices by Country
plt.figure(figsize=(10, 6))
sns.violinplot(x='country', y='high', data=df, palette='pastel')
plt.title('High Prices Distribution by Country')
plt.xlabel('Country')
plt.ylabel('High Price')
plt.xticks(rotation=45)
plt.show()
```
